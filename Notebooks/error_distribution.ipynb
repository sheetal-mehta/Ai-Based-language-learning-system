```python
import pandas as pd
import ast
```

```python
train = pd.read_csv("/content/train.csv")
test= pd.read_csv("/content/test.csv")
```

```python
test.shape
```

```python
train_ann = train['annotations']
```

```python
import numpy as np
def err_dist(ann):
  error_type = {}

  for i in ann:

    if pd.notnull(i):
      i = ast.literal_eval(i)
      for j in i:
          error = j['error_type']
          if error not in error_type:
            error_type[error] = 1
          else:
            error_type[error] = error_type[error]+1

  return error_type
```

```python
train_err_dist = err_dist(train_ann)
```

```python
error_types_all = train_err_dist.keys()
error_types_all
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Sort dictionary by values
sorted_dict = dict(sorted(train_err_dist.items(), key=lambda item: item[1]))

# creating the dataset
data = sorted_dict
courses = list(data.keys())
values = list(data.values())


fig = plt.figure(figsize = (20, 10))

#plt.barh(courses,values)
# creating the bar plot
plt.barh(courses, values, color ='maroon')

plt.xlabel("error types")
plt.ylabel("No. of error")
plt.title("Error type distribution")
plt.show()
```
